```python
train.info()
```

```python
y_train = train["SalePrice"]
del train["SalePrice"]
X_train = train
X_test = test
```

```python
cat_feature = ['Street', 'LotShape', 'LandContour',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'ExterQual', 'ExterCond',
       'Foundation', 'Heating', 'HeatingQC', 'CentralAir', 
       'PavedDrive', 'SaleCondition']
```

```python
for cat in cat_feature:
    dummy = pd.get_dummies(X_train[cat], prefix=cat)
    X_train = pd.concat([X_train, dummy], axis=1)
    del X_train[cat]

for cat in cat_feature:
    dummy = pd.get_dummies(X_test[cat], prefix=cat)
    X_test = pd.concat([X_test, dummy], axis=1)
    del X_test[cat]
```

## (4) 모델 선택 및 학습

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
```

## (5) 평가 / 예측 및 y 값 저장

```python
# 예측하기
pre = knn.predict(X_test)
```
